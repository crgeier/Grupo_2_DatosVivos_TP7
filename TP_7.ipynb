- Agregar categorÃ­a `desconocido` a los valores NaN de las 3 columnas previamente mencionadas.

1) DiscretizaciÃ³n y guardado de bins

````python
# Las 3 variables que nos piden discretizar son las que tienen NAN. A todas vamos a tener que agregarles
# una categorÃ­a que sea "desconocida".
# No me queda claro por quÃ© pide por igual rango e igual cantidad. No sÃ© cÃ³mo
# hacer las 2.

# DiscretizaciÃ³n por igual frecuencia
saved_bins_ph_freq = pd.qcut(df['ph'], q=5, duplicates='drop', retbins=True)[1]
saved_bins_sulfate_freq = pd.qcut(df['Sulfate'], q=5, duplicates='drop', retbins=True)[1]
saved_bins_trihalomethanes_freq = pd.qcut(df['Trihalomethanes'], q=5, duplicates='drop', retbins=True)[1]

# DiscretizaciÃ³n por igual rango
# saved_bins_ph_range = pd.cut(df['ph'], bins=5, duplicates='drop', retbins=True)[1]
# saved_bins_sulfate_range = pd.cut(df['Sulfate'], bins=5, duplicates='drop', retbins=True)[1]
# saved_bins_trihalomethanes_range = pd.cut(df['Trihalomethanes'], bins=5, duplicates='drop', retbins=True)[1]

# Guardo todo en un diccionario con el nombre que nos pidieron... tanto el de
# frecuencia como el de rango.
saved_bins_ph = saved_bins_ph_freq
saved_bins_sulfate = saved_bins_sulfate_freq
saved_bins_trihalomethanes=saved_bins_trihalomethanes_freq
````

2) Guardar los bins con pickle

````python
import pickle

with open('saved_bins_ph.pickle', 'wb') as handle:
    pickle.dump(saved_bins_ph, handle, protocol=pickle.HIGHEST_PROTOCOL)

with open('saved_bins_sulfate.pickle', 'wb') as handle:
    pickle.dump(saved_bins_sulfate, handle, protocol=pickle.HIGHEST_PROTOCOL)

with open('saved_bins_trihalomethanes.pickle', 'wb') as handle:
    pickle.dump(saved_bins_trihalomethanes, handle, protocol=pickle.HIGHEST_PROTOCOL)
````

3) Aplicar discretizaciÃ³n y agregar categorÃ­a "desconocido"

````python
# Aplico discretizaciÃ³n por igual frecuencia a las columnas
df['ph_disc'] = pd.cut(df['ph'], bins=saved_bins_ph_freq, include_lowest=True, duplicates='drop')
df['Sulfate_disc'] = pd.cut(df['Sulfate'], bins=saved_bins_sulfate_freq, include_lowest=True, duplicates='drop')
df['Trihalomethanes_disc'] = pd.cut(df['Trihalomethanes'], bins=saved_bins_trihalomethanes_freq, include_lowest=True, duplicates='drop')

# Reemplazo NaN con categorÃ­a "desconocido"
df['ph_disc'] = df['ph_disc'].cat.add_categories('desconocido').fillna('desconocido')
df['Sulfate_disc'] = df['Sulfate_disc'].cat.add_categories('desconocido').fillna('desconocido')
df['Trihalomethanes_disc'] = df['Trihalomethanes_disc'].cat.add_categories('desconocido').fillna('desconocido')
````

````python
# Verificar que no hay NaN
print("Valores nulos despuÃ©s del tratamiento:")
print(f"ph_disc: {df['ph_disc'].isnull().sum()}")
print(f"Sulfate_disc: {df['Sulfate_disc'].isnull().sum()}")
print(f"Trihalomethanes_disc: {df['Trihalomethanes_disc'].isnull().sum()}")
````

4) Get Dummies (One-Hot Encoding)

````python
# Aplicar get_dummies a las columnas discretizadas
dummies_ph = pd.get_dummies(df['ph_disc'], prefix='ph')
dummies_sulfate = pd.get_dummies(df['Sulfate_disc'], prefix='sulfate')
dummies_trihalomethanes = pd.get_dummies(df['Trihalomethanes_disc'], prefix='trihalomethanes')

# Concatenar con el dataframe original
df_final = pd.concat([df, dummies_ph, dummies_sulfate, dummies_trihalomethanes], axis=1)
````
